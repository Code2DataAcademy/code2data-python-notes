Format

```python
name="Jane"
surname="Tracy"
age=19
print("This is {} {}.She is {}.".format(name,surname,age)) #format string #old
print(f"This is {surname} {name}.She is {age}.") #new
```

```python
print("This is {1} {1}.She is {1}.".format(name,surname)) #format string #old
```

```python
print("{:.4f}".format(3.456778))
```

```python
print(f"3.456789 rounded to the nearest hundredths is {3.456789:.2f}.")
```
